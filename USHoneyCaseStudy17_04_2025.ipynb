**Find the top and bottom 5 states on the basis of production**

```python
# Use the groupby function on production with respect to State.
states_production = df.groupby('state')['production'].sum().reset_index()
states_production
```

```python
states_production.index
```

```python
# Selecting the top 5 best producing states.
top_5_states = states_production.sort_values(by='production',ascending=False).head()
top_5_states
```

```python
# Select the top 5 worst producing states.
bottom_5_states = states_production.sort_values(by='production',ascending=True).head()
bottom_5_states
```

**Find out how much the change in mean average price of Honey from 1995 to 2021**

```python
mean_avg = df.groupby('year')['average_price'].mean().reset_index()
mean_avg
```

**Which was the year when production of Honey in whole US was highest**

```python
total_prod = df.groupby('year')['production'].sum().reset_index()
total_prod
```

```python
max_production_year = total_prod.sort_values(by='production',ascending=False).head(2)
max_production_year
```

```python
highest_production = df.groupby('year')['production'].sum().reset_index()
highest_production.max()
```

**Find out which state was most contributing to make the year 2000 as the max producer**

```python
# production but with respect to year and state.
all_states = df.groupby(['state','year'])['production'].sum().reset_index()
all_states
```

```python
# Now we will select all the rows where the year is 2000.
state_2000_contribution = all_states.loc[(all_states['year']==2000)].reset_index()
```

```python
state_2000_contribution
```

```python
max_contri_state_2000 = state_2000_contribution.sort_values(by='production',ascending=False).head(1)
max_contri_state_2000
```

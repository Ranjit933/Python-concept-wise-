**Find out the relationship between "GDP PER CAPITA" and "Internet Usage" by using Scatter Plot.**

```python
df['Individuals using the Internet (per 100 inhabitants)']
```

```python
sns.scatterplot(data=df,x='Individuals using the Internet (per 100 inhabitants)',y='GDP per capita (current US$)')
plt.title('GDP V/S Internet Users')
plt.xlabel('GDP')
plt.ylabel('Internet Users')
plt.show()
```

**Plot a bar chart showing top 10 countries by GDP Per Capita**

```python
top_10_countries = df.sort_values(by='GDP per capita (current US$)',ascending=False).head(10)
top_10_countries
```

```python
plt.bar(top_10_countries['country'],top_10_countries['GDP per capita (current US$)'])
plt.title('Top 10 Highest GDP Contributing Countries')
plt.ylabel('GDP')
plt.xticks(rotation=90)
plt.show()
```

```python
#PCA:
#1. prepare the data

X=df.iloc[:,:-1]
Y=dataset.target
```

```python
X
```

```python
Y
```

```python
#2. Standardization
# scale data in a way that mean=0, std dev=1

from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
S=scaler.fit_transform(X)
```

```python
S
```

```python
pd.DataFrame(S)
```

```python
#3. Covariance Matrix:helps PCA understand how your data features are related, so it can reduce the dimensions while keeping the most important info

ST=S.T # Transpose: changes rows to columns and vice versa
pd.DataFrame(ST)
```

```python
cov_mat=np.cov(ST) #how 2 variables change together
pd.DataFrame(cov_mat)
```

```python
#4. Eigen values and Eigen vectors
#Eigen vectors: direction that stays the same during a transformation(only length changes)
#Eigen values: How much that direction is stretched or shurnk
```

```python
eig_values,eig_vectors=np.linalg.eig(cov_mat)
```

```python
pd.DataFrame(eig_values)
```

```python
pd.DataFrame(eig_vectors)
```

```python
Tot_Eig=sum(eig_values)
```

```python
Tot_Eig
```

```python
sorted(eig_values, reverse=True)
```

```python
var_exp=[]
for x in sorted(eig_values,reverse=True):
  exp=(x/Tot_Eig)*100
  var_exp.append(exp)
var_exp
```

```python
#Cumulative Variance: helps decide how many components to keep

CV=np.cumsum(var_exp)
CV
```

```python
CVD=pd.DataFrame(CV)
CVD.head(30)
```

```python
#APPLY PCA

from sklearn.decomposition import PCA

pca=PCA(0.95) #This tells PCA to retain enough components such that 95% of the variance in the data is present
X_pca=pca.fit_transform(X)
```

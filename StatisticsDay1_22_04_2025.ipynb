# **Statistics:**
It is a branch of mathematics that deals with collection of data. It help us analyse the data, presentation of data , organization of data in such a way that we can extract some meaningful information.

**Understand the Data:** Identify the pattern, relationship.

**Prepare the Data:** Help to clean the data.

# **Statistics is divided into two Categories**
**Descriptive Statistics:** Describe the main features of data instead of analysing completely.

1.   *Measure of Central Tendency*
2.   *Measure of Dispersion*

# **Measure of Central Tendency**

It is used to indicate where does the centre of our data lies.
1.  Mean
2.  Mode
3.  Median

1.  Population
2.  Sample

# **Measure of Dispersion**
It is used to indicate how much our data is spread out/dispersed in all the direction.

1.  Range
2.  Variance
3.  Standard Deviation
4.  Percentile
5.  Quartile
6.  Outliers
7.  Co-relation
8.  Normal Distribution
9.  Skewness

**Inferential Statistics (Prediction and Conclusion):** It is a branch of Statistics that deals with making prediction and drawing conclusion on sample data collection.
1.  **Hypothesis Testing:** Null Hypothesis, Alternate Hypothesis
2.  **Regression Analysis:**  Dependent Variable , Independent Variable

# *Mode: Most frequent values present in our data/ Maximum times repeated element in our data.*

```python
ele = [3,4,2,1,3,1,2,4,3,4,1,3]
mode_ele = stat.mode(ele)    #In numpy we don't have mode.
print(mode_ele)
```

```python
ele_same = [1,4,2,1,3,3,2,4,3,4,1,3,1]
mode_ele_same = stat.mode(ele_same)    #In numpy we don't have mode.
print(mode_ele_same)
# If more than one element we are getting as the final mode, the element that occuring first
# will become our mode.
```

```python
mul_mode = stat.multimode(ele_same)
print(mul_mode)
```

```python
data = [10,20,30,40,50]
# 10 - 1
# 20 - 1
# 30 - 1
# 40 - 1
# 50 - 1
stat.multimode(data)
```

# *Population: Entire Data*
# *Sample: A subset of entire Population, selecting few random people from entire population*

```python
import random
```

```python
data_elements = [10,150,200,170,145,125,190,65,74,38,29,30] # Entire Population
sample_results = random.sample(data_elements,6)  #Sample
sample_results
```

```python
# Random State.
names = ['Hiba','Vishnu','Ankit','Deepak','Pooja','Harish','Vaishnavi','Lakshmi','Charan']
print(names)
```

```python
# Out of these 9 (Entire Data/Population) I want to select 4 people randomly.
# These 4 people will be our sample.
```

```python
sample_names = random.sample(names,4)
sample_names
```

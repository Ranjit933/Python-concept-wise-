# **Statistics:**
It is a branch of mathematics that deals with collection of data. It help us analyse the data, presentation of data , organization of data in such a way that we can extract some meaningful information.

**Understand the Data:** Identify the pattern, relationship.

**Prepare the Data:** Help to clean the data.

# **Statistics is divided into two Categories**
**Descriptive Statistics:** Describe the main features of data instead of analysing completely.

1.   *Measure of Central Tendency*
2.   *Measure of Dispersion*

# **Measure of Central Tendency**

It is used to indicate where does the centre of our data lies.
1.  Mean
2.  Mode
3.  Median

1.  Population
2.  Sample

# **Measure of Dispersion**
It is used to indicate how much our data is spread out/dispersed in all the direction.

1.  Range
2.  Variance
3.  Standard Deviation
4.  Percentile
5.  Quartile
6.  Outliers
7.  Co-relation
8.  Normal Distribution
9.  Skewness

**Inferential Statistics (Prediction and Conclusion):** It is a branch of Statistics that deals with making prediction and drawing conclusion on sample data collection.
1.  **Hypothesis Testing:** Null Hypothesis, Alternate Hypothesis
2.  **Regression Analysis:**  Dependent Variable , Independent Variable

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as stat
import random
```

```python
age = [18,25,30,45,33,-190,67,78,89,93,250,76,99,271]
age
```

```python
plt.boxplot(age)
plt.show()
```

image.png

```python
# sns.violinplot(age)
# plt.show()
```

```python
# Step 1: Calculate Q1 and Q3
Quartile_1 = np.percentile(age,25)
Quartile_3 = np.percentile(age,75)
print(Quartile_1 ,Quartile_3)
```

```python
age
```

```python
# Step 2: Calculate IQR (Inter Quartile Range)
# IQR = Q3 - Q1
IQR = Quartile_3 - Quartile_1
IQR
```

```python
np.percentile(age,50)
```

```python
# Step 3: Calculate Upper Limit & Lower Limit
lower_limit = Quartile_1 - 1.5*IQR  #lower whisker
upper_limit = Quartile_3 + 1.5*IQR  #upper whisker
```

```python
print(lower_limit,upper_limit)
```

```python
# Step 4: Remove Outliers
# We are going to remove all the values that are smaller than to lower limit.
# We are going to remove all the values that are greater than to upper limit.
```

```python
print(age)
```

```python
for x in age:
    if x < lower_limit or x > upper_limit:
        age.remove(x)
```

```python
plt.boxplot(age)
plt.show()
```

```python
sns.violinplot(age)
plt.show()
```

# **Probability**

It is a measure of how likely an event is going to occur.

1.  **Experiment:** Any process or task. Examples: Flipping a coin , Rolling a dice , drawing a card from a deck.

2.  **Event:** Event is a specific or group of outcomes from an experiment.

*  *Simple Event:* Means 1 specific outcome. Ex: Flipping a coin and getting a tail.

*  *Compound Event:* Means multiple outcomes. Rolling a dice and asking for even numbers (2,4,6).

**Outcome:** It is like a result of single trial of experiment. Ex: When you are going to flip a coin the outcome can be "Head" or "Tail".

image.png

```python
# What will be the probability of getting a number less than 4 while rolling a dice.

# Number of Fav Outcome: 3 (1,2,3)
# Total No. of Outcome: 6 (1,2,3,4,5,6)
# P(<4) = 3/6 => 1/2

# Number of cards in a deck = 52.
# What will be probability of getting a Face Card?
# King , Queen and Jack
# Favorable Outcome: 12
# Total No. of Outcome: 52
# P => 12/52 => 3/13
```

```python
# What is the probability of drawing a red card from a deck.
# Fav Outcome  = 26 (Either it can be heart or diamond)
# Total Outcome = 52
# Prob = 26/52 => 1/2.


# What is the probability of getting a multiple of 3 on rolling a dice?
# Fav Outcome = 2 (3,6)
# Total Outcome = 6 (1,2,3,4,5,6)
# Prob = 2/6 => 1/3.
```

# **Marginal Probability**

It is the probability of single event happening without considering any other event.

*Example:* You are having a bag full of balls, 5 red balls and 3 blue balls. What are the chances of marginal probability of picking a red ball out of total balls?

```python
# Favorable Outcome = 5 (5 red balls)
# Total no. of Outcome = 8 (Total Number of Balls)
# Prob => 5/8
```

# *Joint Probability*

It is the probability of two or more events happening at the same time. It's just like *What are the chances of this and that happening at the same time.*

*Example:* If you roll 2 dice at the same time what will be the probability of both the dice showing 5.

```python
# P(A and B) = P(A) x P(B)
# P(A) = Dice_1
# P(B) = Dice_2
# P(A) = 1/6
# P(B) = 1/6
# P(5) = P(A)xP(B)
# P(5) = 1/36
```

*Example:* What is the probability of getting a Head on both the coins when two coins are tossed together.

```python
# P(A) = Coin_1
# P(B) = Coin_2
# P(A) = 1/2
# P(B) = 1/2
# P(H) = 1/4

# Number of Fav Outcome = 1 (H,H)
# Total No. of Outcomes = 4((H,H),(H,T),(T,H),(T,T))
# Prob = 1/4.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/cars.csv',skipinitialspace=True)
df
```

```python
df.dropna(inplace=True)   # dropna is used to remove null values from our dataset.
```

**Create scatter plots for the following columns**

1.   mpg-hp
2.   mpg-weight
3.   hp-mpg
4.   hp-time-to-60  
**with respect to each brand. Make sure all the plots appear in the same plane with x-labels and y-labels evenly spaced.**



```python
plt.scatter(df['mpg'],df['hp'])
plt.xlabel('Miles Per Gallon')
plt.ylabel('Horse Power')
plt.title('MPG V/S HP')
plt.show()
```

```python
sns.scatterplot(x=df['mpg'],y=df['hp'],hue=df['brand']) # Hue parameter gives different colors to data point with respect to column.
plt.xlabel('Miles Per Gallon')
plt.ylabel('Horse Power')
plt.title('MPG V/S HP')
plt.show()
```

```python
fig = plt.subplots(2,2,figsize=(20,10))

plt.subplot(2,2,1)
sns.scatterplot(x=df['mpg'],y=df['hp'],hue=df['brand'])
plt.xlabel('Miles Per Gallon')
plt.ylabel('Horse Power')
plt.title('MPG V/S HP')

plt.subplot(2,2,2)
sns.scatterplot(x=df['mpg'],y=df['weightlbs'],hue=df['brand'])
plt.xlabel('Miles Per Gallon')
plt.ylabel('Weight LBS')
plt.title('MPG V/S WeightLBS')

plt.subplot(2,2,3)
sns.scatterplot(x=df['hp'],y=df['mpg'],hue=df['brand'])
plt.xlabel('Horse Power')
plt.ylabel('Miles Per Gallon')
plt.title('HP V/S MPG')

plt.subplot(2,2,4)
sns.scatterplot(x=df['hp'],y=df['time-to-60'],hue=df['brand'])
plt.xlabel('Horse Power')
plt.ylabel('Time to 60')
plt.title('HP V/S Time_to_60')

plt.show()
```

**Create bar plots for the following columns:**


1.   cylinders-hp
2.   cylinders-mpg
3.   cylinders-weight-
4.   cylinders-time-to-60
**Make sure all the plots appear in the same plane with x-labels and y-labels evenly spaced.**



```python
fig = plt.subplots(2,2,figsize=(18,12))

plt.subplot(2,2,1)
sns.barplot(x=df['cylinders'],y=df['hp'],hue=df['brand'])
plt.xlabel('Cylinders')
plt.ylabel('Horse Power')
plt.title('Cylinders V/S Horse Power')

plt.subplot(2,2,2)
sns.barplot(x=df['cylinders'],y=df['mpg'],hue=df['brand'])
plt.xlabel('Cylinders')
plt.ylabel('Miles Per Gallon')
plt.title('Cylinders V/S MPG')

plt.subplot(2,2,3)
sns.barplot(x=df['cylinders'],y=df['weightlbs'],hue=df['brand'])
plt.xlabel('Cylinders')
plt.ylabel('WeightLBS')
plt.title('Cylinders V/S WeightLBS')

plt.subplot(2,2,4)
sns.barplot(x=df['cylinders'],y=df['time-to-60'],hue=df['brand'])
plt.xlabel('Cylinders')
plt.ylabel('time-to-60')
plt.title('Cylinders V/S time-to-60')

plt.show()
```

# Is a Sample Different from the Population?
Problem: A school claims that the average height of students is 170 cm. A random sample of 50 students has an average height of 165 cm, with a population standard deviation of 10 cm. At a 5% significance level, can we conclude that the students’ height differs from the population mean?

```python
# Given Data
sample_mean = 165
population_mean = 170
population_std = 10
sample_size = 50
alpha = 0.05

# Null Hypothesis: mu = 170
# Alternate Hypothesis: mu != 170
# z test (Two Tailed)

# Calculate the z score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the p_value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Print the result
print("Z-Score:", z_score)
print("P-Value:", p_value)

# Decide on the hypothesis
if p_value < alpha:
    print("Reject the null hypothesis. The students' height differs from the population mean.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the students' height and the population mean.")

```

# Is the Sample Mean Smaller?
Problem:
The national average score in math is 75. A school wants to test if their students' average score is less than 75. A sample of 36 students has an average score of 72, and the population standard deviation is 8. Use a appropriate test at a 5% significance level.

```python
# Given Data
sample_mean = 72
population_mean = 75
population_std = 8
sample_size = 36
alpha = 0.05

# Null Hypothesis: mu = 75
# Alternate Hypothesis: mu < 75
# z test (One Tailed - Left)

# Calculate the z score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the p-value
p_value = stats.norm.cdf(z_score)

# Print the result
print("Z-Score:", z_score)
print("P-Value:", p_value)

# Decide on the hypothesis
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

# Before and After Treatment
Problem:
A fitness trainer measures the weight of 8 people before and after a 4-week training program. Their weights (in kg) are as follows:

*   Before: [80, 85, 78, 90, 95, 88, 76, 82]
*   After: [78, 84, 76, 88, 94, 87, 74, 81]

Can we conclude at a 5% significance level that the training program had a significant effect on their weight?

```python
# Given Data
before = [80, 85, 78, 90, 95, 88, 76, 82]
after = [78, 84, 76, 88, 94, 87, 74, 81]
alpha = 0.05

# Perform the paired t-test
t_statistic, p_value = stats.ttest_rel(before, after)

# Print the result
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Decide on hypothesis
if p_value < alpha:
    print("Reject the null hypothesis. The training program had a significant effect on their weight.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in weight before and after the training program.")
```

* 4.9, 5.1, 4.8, 5.0, 5.2, 4.9, 5.0, 5.1, 5.0, 4.8

Can we conclude at a 5% significance level that the sample mean is different from 5 mm?

# ANOVA tests

ANOVA (Analysis of Variance) is a statistical technique used to determine if there are statistically significant differences between the means of three or more independent groups.


1. One-Way ANOVA Example
Problem: Suppose a data scientist wants to compare the effectiveness of three marketing channels (TV, Social Media, and Radio) in terms of average monthly sales.

```python
import pandas as pd
import scipy.stats as stats

# Create a sample dataset
data = {
    'Sales': [200, 220, 215, 300, 305, 290, 180, 185, 190, 260, 270, 265],
    'Channel': ['TV', 'TV', 'TV', 'Social Media', 'Social Media', 'Social Media',
                'Radio', 'Radio', 'Radio', 'Radio', 'Radio', 'Radio']
}
df = pd.DataFrame(data)
df
```

```python
# Perform the One Way ANOVA
anova_statistic, p_value = stats.f_oneway(
    df[df['Channel'] == 'TV']['Sales'],
    df[df['Channel'] == 'Social Media']['Sales'],
    df[df['Channel'] == 'Radio']['Sales']
)

# Print the results
print("ANOVA Statistic:", anova_statistic)
print("P-Value:", p_value)
```

```python
# Decide on hypothesis
alpha = 0.05 # Assumed 5% significance
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in sales between the marketing channels.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in sales between the marketing channels.")
```

# Two Way ANOVA

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Sample dataset with two categorical factors (Region and Channel)
data = {
    'Sales': [300, 310, 305, 350, 345, 340, 200, 210, 205, 250, 245, 240],
    'Region': ['North', 'North', 'North', 'South', 'South', 'South',
               'North', 'North', 'North', 'South', 'South', 'South'],
    'Channel': ['TV', 'TV', 'TV', 'TV', 'TV', 'TV',
                'Radio', 'Radio', 'Radio', 'Radio', 'Radio', 'Radio']
}
df = pd.DataFrame(data)
df
```

```python
# Define the model for two way ANOVA
model = ols('Sales ~ C(Region) + C(Channel) + C(Region):C(Channel)', data=df).fit()

# Perform the two way ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print(anova_table)
```

```python
# Decide on Hypothesis
alpha = 0.05 # Assumed 5% significance
if anova_table['PR(>F)']['C(Region)'] < alpha:
    print("Reject the null hypothesis. There is a significant effect of  Region on Sales.")

elif anova_table['PR(>F)']['C(Channel)'] < alpha:
    print("Reject the null hypothesis. There is a significant effect of Channel on Sales.")

elif anova_table['PR(>F)']['C(Region):C(Channel)'] < alpha:
    print("Reject the null hypothesis. There is a significant effect of interaction of Region and Channel on Sales.")

else:
    print("Fail to reject the null hypothesis. There is no significant interaction effect between Region and Channel.")
```

1. Chi-Square Test of Independence

A company wants to check if the customer's payment method (Credit Card, PayPal, Cash) is associated with the region (North, South). The company collects data from customers across different regions.

```python
from scipy.stats import chi2_contingency

# Sample dataset
data = {'Credit Card': [30,50], 'Pay Pal': [20,40], 'Cash':[50,30]}
df = pd.DataFrame(data, index=['North', 'South'])
df
```

```python
# Perform the Chi-Square test - Test of Independance
chi2, p, dof, expected = chi2_contingency(df)

# Print the results
print("Chi-Square Statistic:", chi2)
print("P-Value:", p)
print("Degrees of Freedom:", dof)
print("Expected Values", expected)
```

```python
# Decide on hypothesis
alpha = 0.05 # Assumed 5% significance
if p < alpha:
    print("Reject the null hypothesis. There is a significant association between payment method and region.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between payment method and region.")
```

2. Chi-Square Goodness-of-Fit

Problem: A retailer expects that 30% of its customers will choose Product A, 50% Product B, and 20% Product C. After running a survey, they get the following responses:


*   Product A: 25 customers
*   Product B: 45 customers
*   Product C: 30 customers


Does the observed distribution match the expected proportions?

```python
# Observed Data
observed = [25, 45, 30]

# Expected Data
expected_prop = [0.3, 0.5, 0.2]
total_customers = sum(observed)

# Calculate the count of Expected People
expected = [i * total_customers for i in expected_prop]

#Print the results
print("Observed:", observed)
print("Expected:", expected)
print("Expected Proportions", expected_prop)
```

```python
# Perform the chi Square test - Goodness of fit
chi2, p = stats.chisquare(f_obs=observed, f_exp=expected)

# Print the results
print("Chi-Square Statistic:", chi2)
print("P-Value:", p)

# Decide on hypothesis
alpha = 0.05 # Assumed 5% significance
if p < alpha:
    print("Reject the null hypothesis. The observed distribution does not match the expected proportions.")
else:
    print("Fail to reject the null hypothesis. The observed distribution matches the expected proportions.")
```

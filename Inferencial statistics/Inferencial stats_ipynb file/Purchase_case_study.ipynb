#a> it was observed that the avg purchase made by the men of age group 18-25 was 10000.is it still the same

```python
#one sample ttest
```

null:mean will be 10k

alt:mean is not equal to 10k

```python
#data----df

df1 = df[(df['Age'] == 1) & df['Gender'] == 1]
df1
```

```python
df1['Purchase'].mean()
```

```python
sample = df1.sample(3600,random_state = 0)
```

```python
sample
```

```python
from scipy.stats import ttest_1samp
a_mean = 10000
```

```python
t_stat,p_value = ttest_1samp(sample['Purchase'],a_mean)
```

```python
print(p_value)
```

as  the p_val is less than 0.05 we will reject the null hypo
so,the mean is not equal to 10k

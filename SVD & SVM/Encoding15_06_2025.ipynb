**Encoding:** is the concept to convert our categorical data(columns) into numerical columns.


1.   Label Encoding
2.   Ordinal Encoding
3.   One hot Encoding



```python
import pandas as pd
```

```python
dummy ={
    'Color':['Red','Green','Blue','Black','Red','Green'],
    'Size':['Small','Large','Medium','Medium','Large','Small'],
    'Marital_Status':['Married','Unmarried','Married','Relationship','Unmarried','Relationship'],
    'Gender':['Male','Female','Male','Male','Female','Female']
}
df = pd.DataFrame(dummy)
df
```

# **Label Encoding**

**It uses alphabetical order to encode the data**

```python
df['Color']
```

image.png

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Color'] = le.fit_transform(df['Color']) # This fxn will convert our categorical column to numerical
```

```python
df['Color']
```

```python
df
```

```python
df['Size']
```

# **Ordinal Encoding**

When we want to decide the order of encoding for our categorical data.

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Small','Medium','Large']])
df['Size'] = oe.fit_transform(df[['Size']])
```

```python
df["Size"].astype(int)
```

# **OneHot Encoding**

```python
df['Marital_Status']
```

```python
df = pd.get_dummies(df,columns=['Marital_Status'])
```

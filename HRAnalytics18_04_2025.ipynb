**Do number of projects effects employees leaving the company?**

```python
# Can we create a countplot? Can I create a count plot with respect to 'left' column?
```

```python
sns.countplot(data=df,x='numberOfProjects')
plt.show()
```

```python
sns.countplot(data=df,x='numberOfProjects',hue='left')
plt.show()
```

```python
# Conclude: None of the employees who were working on 7 projects stayed in the company. All of them left:
# Improvement -> Reduce the number of projects. Reduce work stress.

# Conclude: Employees with 2 projects are also leaving the company. Employees with min and max projects are leaving the company
# Improvement -> Divide number of projects between 3 and 5.
```

```python
df[(df['numberOfProjects']==3) & (df['left']==1)].shape  # Only 38 employees left the company while working on 3 number of projects.
```

**Do timespent in company(experience) effects employee in leaving the company?**

```python
sns.countplot(data=df,x='timeSpent.company',hue='left')
plt.show()
```

```python
# Insights
# Employees who spent 3,4,5,6 years in company are leaving the most.
# Employees with 7 and 7+ years of exp are not leaving the company at all.

# Suggestions
# Promotion after 3 years -> Salary Increase -> Long term Incentive -> Vacation
# Hike in Salary and Position
# Giving them higher position and interaction with interns.
# Assign employees with 3,4,5,6 exp with 3 to 4 number of projects.
```

**Does the Department effects employees leave the company?**

```python
sns.countplot(data=df,x='dept',hue='left',palette=['red','black'])
plt.xticks(rotation=45)
plt.show()
```

```python
# Employees from sales, technical and support are leaving the company most.

# Suggestions:
# Make good work life balance, more incentive to Sales people on achieving the goals.
# Give option to switch department.
# Have better communication with these department employees.
```

**Find out the relation between department and Salary**

```python
sns.countplot(data=df,x='dept',hue='salary',palette=['blue','green','black'])
plt.xticks(rotation=90)
plt.show()
```

```python
# Insights:
# In Sales , technical and support people are having most low salaries.
# Managament is having good Salary distribution (Mgmt dept is having highest % of high salary)

# Suggestions:
# Salary distribution can be improved
# Promotion in Sales, technical and support
# More incentive to Sales people
# On the basis of projects you can provide Salary to technical people.
```

**Does Salary affect employees leaving the company?**

```python
ax = sns.countplot(data=df,x='salary',hue='left')
for x in ax.containers:
    ax.bar_label(x)
plt.show()
```

```python
# Insights:
# Low salary emps are leaving the company the most.
# High attrition in low and medium salary employees

# Suggestions:
# Give timely promotion to employees
# Provide salary according to the Market Standards
# Yearly Increment.
# Recognize the work efforts and provide rewards.
```

**Check for outliers, if there are outliers in our data remove them**

```python
for i in df.columns:
    if df[i].dtype=='object':
        continue
    sns.boxplot(data=df,x=i)
    plt.title(i)
    plt.show()
```

```python
for x in df.columns:
    if df[x].dtype=='object':
        continue

    Q1 = df[x].quantile(0.25)
    Q3 = df[x].quantile(0.75)
    IQR = Q3-Q1
    LL = Q1 - 1.5*IQR
    UL = Q3 + 1.5*IQR
    df = df[(df[x]>=LL) & (df[x]<=UL)]
```

```python
for x in df.columns:
    if df[x].dtype!='object':
        plt.boxplot(df[x])
        plt.show()
```

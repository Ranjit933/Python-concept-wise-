#### Applying filters on data set

```python
df1.head()
```

```python
cond =  (df1['origin'] == 1) # extract the data where origin value matches with 1 
cond
```

```python
# cond consists of True / false as output of boolean operator 
df1[cond] # Extract the rows where value is True
```

```python
cond =  (df1['origin'] == 1) & (df1['year'] > 70) # extract the data where origin value matches with 1 

df1[cond]
```

```python
code = (df1['origin']==1) & (df1['year'] > 70)
df1[code]
```

```python
df1[(df1['origin']==1) & (df1['year'] > 70)]['mpg'].mean()
```

```python
df2 = df1[(df1['origin']==1) & (df1['year'] > 70)]
df2['mpg'].mean()
```

```python
select 
    sum(mpg) 
from df1 
groupby origin 
# take df1 and partition the data based on unique values in origin column and then aggregate mpg column sum / mean / min / max 
```

```python
df1.groupby('origin').agg({'mpg': 'mean', 'acceleration': 'min'})
```

```python
df1[['origin','mpg','acceleration']].groupby('origin').agg(['min','mean'])
```

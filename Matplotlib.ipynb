```python
import matplotlib.pyplot as plt
A = [21,22,23,24,25]
B = [22,24,21,25,23]
C = [1,2,3,4,5]
```

```python
help(plt.scatter)
```

```python
plt.scatter(C,A) # c -> in x axis , A --> y axis
```

```python
plt.scatter(C,A, marker = '.',color = 'Red')

# marker : D- diamond , H - Hexagonal, 'o' - circle ,'.' - points
plt.xlabel('Months')
plt.ylabel('no.of products sold')
plt.title('A-Sales Performance')
plt.grid() # to enable background lines / reference lines 

plt.show() #use plt.show() to hide <matplotlib.collections.PathCollection at 0x23d0cc2d110>
```

```python
plt.scatter(C,A, marker = 'D',color = 'Red')
plt.scatter(C,B, marker = 'h', color = 'black')
# marker : D- diamond , H - Hexagonal, 'o' - circle ,'.' - points
plt.xlabel('Months')
plt.ylabel('no.of products sold')
plt.title('A-Sales Performance')
plt.grid() # to enable background lines / reference lines 
plt.legend(['Person A','Person B']) # display the info about graph 1 and graph2 
plt.show() #use plt.show() to hide <matplotlib.collections.PathCollection at 0x23d0cc2d110>
```

```python
plt.subplot(121) # 1 row 2 columns 1st graph
plt.scatter(C,A, marker = 'D',color = 'Red')
plt.xlabel('Months')
plt.ylabel('no.of products sold')
plt.title('A-Sales Performance')
plt.grid() # to enable background lines / reference lines 
plt.legend(['Person A']) # display the info about graph 1 and graph2 
plt.subplot(122) # 1 row 2 columns 2nd graph
plt.scatter(C,B, marker = 'h', color = 'black')
plt.xlabel('Months')
plt.ylabel('no.of products sold')
plt.title('B-Sales Performance')
plt.grid() # to enable background lines / reference lines 
plt.legend(['Person B']) # display the info about graph 1 and graph2 
plt.show()
```

```python
plt.subplot(221) # 1 row 2 columns 1st graph
plt.scatter(C,A, marker = 'D',color = 'Red')
plt.xlabel('Months')
plt.ylabel('no.of products sold')
plt.title('A-Sales Performance')
plt.grid() # to enable background lines / reference lines 
plt.legend(['Person A'], loc = 'upper left') # display the info about graph 1 and graph2 


plt.subplot(222) # 1 row 2 columns 1st graph
plt.plot(C,A, linestyle = '--',color = 'Red')
plt.xlabel('Months')
plt.ylabel('no.of products sold')
plt.title('A-Sales Performance')
plt.grid() # to enable background lines / reference lines 
plt.legend(['Person A']) # display the info about graph 1 and graph2 


plt.subplot(223) # 1 row 2 columns 2nd graph
plt.scatter(C,B, marker = 'h', color = 'black')
plt.xlabel('Months')
plt.ylabel('no.of products sold')
plt.title('B-Sales Performance')
plt.grid() # to enable background lines / reference lines 
plt.legend(['Person B']) # display the info about graph 1 and graph2 

plt.subplot(224) # 1 row 2 columns 2nd graph
plt.plot(C,B, color = 'black')
plt.xlabel('Months')
plt.ylabel('no.of products sold')
plt.title('B-Sales Performance')
plt.grid() # to enable background lines / reference lines 
plt.legend(['Person B']) # display the info about graph 1 and graph2 

plt.subplots_adjust(wspace = 0.3, hspace = 0.5)
plt.show()
```

```python
fig,axs = plt.subplots(2,2)
axs[0,0].scatter(C,A)
axs[0,1].plot(C,A)
axs[1,0].scatter(C,B)
axs[1,1].plot(C,B)
plt.show()
```

```python
# Now encode these 2 object and 1 boolean column.
# round_winner   bomb_planted   map

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['map'] = le.fit_transform(df['map'])
df['round_winner'] = le.fit_transform(df['round_winner'])
df['bomb_planted'] = le.fit_transform(df['bomb_planted'])
```

```python
df.info()
```

```python
# Store all the independent columns in X
# Dependent column in y
# Divide your data into 80:20 ratio (Train:Test) and build a logistic regression model.
# Check for accuracy score.
# 8:05 AM.

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X = df.iloc[:,:-1]
y = df['round_winner']
X
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
# Our model took too much time to execute
# 96 features means high dimension
# And even after using max_iter the model was not able to cover the entire data.
```

```python
# PCA
# LDA works a lot better with classification problemns so we will go with LDA
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
# Before applying LDA we need to scale our data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
lda.fit(x_train,y_train)
```

```python
lda.transform(x_test)
```

```python
pd.DataFrame(x_train)
```

```python
y_train
```

```python
# Now we need to decide how many column we need to keep.

# min(independent_features, number_of_classes_in_target-1)

# min(96,1) -> 1
# We cannot convert our 96 columns into 1 single column (Inefficient)
```

```python
# To select the number of components we will calculate the lda_coefficient

lda_coef = np.exp(abs(lda.coef_))
lda_coef
```

```python
lda_coef.shape
```

```python
lda_coef = lda_coef.flatten() #Converting our 2d array to 1d array
```

```python
lda_coef
```

```python
X
```

```python
feature_names = X.columns
```

```python
feature_names
```

```python
di = {
    'Features':feature_names,
    'Impact':lda_coef
}
df_feature_imp = pd.DataFrame(di)
df_feature_imp
```

```python
top_20 = df_feature_imp.sort_values(by='Impact',ascending=False).head(20)
# Randomly selecting topo 20 features
top_20
imp_col = top_20.index
imp_col
```

```python
x_train = x_train[:,imp_col]
```

```python
pd.DataFrame(x_train)
```

```python
x_test = x_test[:,imp_col]
x_test
```

```python
# Again build the logisitic regression and check how your model is working.
final_model = LogisticRegression()
final_model.fit(x_train,y_train)
```

```python
pred = final_model.predict(x_test)
```

```python
accuracy_score(pred,y_test)
```

```python
# Before applying LDA -> 75% accuracy
# After applying LDA -> 75% accuracy
# After LDA we reduced the number of dimensions from 96 to 20
# and still the model is working with same efficiency.

# Homework:
# Build 2 new models
# Decision Tree
# Random Forest
```

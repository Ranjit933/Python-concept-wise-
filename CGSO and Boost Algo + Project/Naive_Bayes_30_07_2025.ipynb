```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Divide the data into X and y and perform train_test_split

X = df.drop(columns=['target'])
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=32)
```

```python
# Gaussian Naive Bayes:
# Use: When your features are continuous numbers (real values)
# Example: Age, height, weight, income, temperature
# Anything that can have decimal values.
# How it works:
# Assumes the values follow a bell-shaped (normal) curve and calculate probabilites
# using mean and standard deviation

# Use Case: Predicting someone has diabetes bases on blood pressure, glucose etc.




# Multinomial Naive Bayes:
# When your features are count of things:

# Example: Numbers of times a word appears in a document, frequncy of clicks
# How it works:
# Use frequencies, more occurance = higher impact
# Use Case: Spam detection etc.


# Bernoulli Naive Bayes:
# When your features are binary (0/1, Yes/No)

# Examples: Word present or not
# Fever: Yes/No

# How it works:
# It will check the presence or absence of a feature. (Doesn't care about how many times it occured)
# Classifying a mail as spam based on whether certain words exist (Not their count)


# Categorical Naive Bayes
# USed when your features are categorical labels
# Example: Color, Marital Status etc

# How it works:Uses probability of each category level given the class
# Predicting if someone will buy a product based on city, gender, class_type

```

```python
X
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import CategoricalNB
```

```python
gnb_model = GaussianNB()
cnb_model = CategoricalNB()
```

```python
gnb_model.fit(x_train,y_train)
cnb_model.fit(x_train,y_train)
```

```python
y_pred = gnb_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

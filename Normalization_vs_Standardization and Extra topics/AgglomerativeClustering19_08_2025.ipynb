```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/cars (1) (1) (1).csv')
df
```

```python
df.info()
```

```python
df_copy = df.drop(['model'],axis=1)
```

```python
df_copy
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage


# Single Linkage -> Min
# Complete Linkage -> Max
# Average Linkage -> Avg
# Wards Linkage -> We create cluster by taking the min variance.
```

```python
z = linkage(df_copy,method='ward')  #  Wards link -> We create clusters by taking min variance

```

```python
plt.figure(figsize=(15,7))
dendrogram(z,labels=np.arange(1,len(df_copy)+1))  # 1 - stopping -> 33
plt.title('Hierarchical Clustering')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

**1. K-Means Clustering**
Use when:

You already know (or can guess) the number of clusters.

Data is large and you want a fast algorithm.

Clusters are expected to be spherical and similar in size.

*Example:* Grouping customers into 34 segments for marketing.


**2. Hierarchical Clustering**
Use when:

You dont know the number of clusters.

Dataset is small to medium sized.

You want to see cluster formation step by step using a dendrogram.

*Example:* Organizing books into topics without knowing how many categories exist.


**3. KNN (K-Nearest Neighbors)**
Use when:

Its a classification or regression problem, not clustering.

You want to make predictions based on similar past data points.

Works well with small datasets and clear neighborhood patterns.

*Example:* Classifying a new patient as diabetic or non-diabetic based on past patients with similar records.



**4. GMM (Gaussian Mixture Model)**
Use when:

Clusters are not spherical (can be oval/elliptical).

You want soft clustering (probability of belonging to each cluster, not hard assignment).

Useful in high-dimensional continuous data.

*Example:* Speaker recognition (each persons voice is modeled as a Gaussian distribution).


**5. Dendrogram & Agglomerative**
Use when:

You want to decide the number of clusters visually.

Need to understand relationships between data points (who merges with whom).

Dataset is not too large (since its slower than K-Means).

*Example:* Understanding genetic similarity between species.



* K-Means  Fast, known number of clusters, spherical shapes.

* Hierarchical/Dendrogram  Dont know clusters, want to visualize merging.

* KNN  Classification/regression, not clustering.

* GMM  Flexible shapes, soft clustering, probability-based.

## ***Introduction to OLS (Ordinary Least Squares)***

**OLS is a technique used to find the best fit line (or hyperplane) in linear regression by minimizing the sum of squared differences (errors) between actual and predicted data points**

**It minimizes the squared difference between observed and predicted values**

**Imagine you have data points scattered on a graph. OLS heps you draw a straight line through these points that best respresents the trend**

## ***Why is OLS used***


**OLS is used because:**

**To predict outcomes:** If you know one variable (e.g., hours studied), OLS helps predict another variable (e.g., exam score).

**To find relationships:** It identifies how changes in one factor (like temperature) affect another (like umbrella sells).

## ***When is OLS used in ML***

**OLS forms the foundation for:**

**Linear Regression Models:** IN supervised learning, linear regression uses OLS to compute coefficents for the relationship between features and the target.

**Model Validation:** OLS assumptions (like linearity) help evaluate how well the model fits the data.

## OLS FORMULA
$$
{\beta} = (X^T X)^{-1} X^T y
$$
* ${\beta}$: Represents the coefficients of the regression model.
* **X**: The Design Matrix (Features)
* $X^T$: Transpose of X
* $X^T X$: The feature covariance matrix (to summarize the relationship between our features)
* $(X^T X)^{-1}$: The inverse of $X^T X$ to undo the effect of multiply so we can solve for the coefficient.
* **y**: Target Vector / Dependent Variable
* $X^T y$: Feature Target Relationship

**Ordinary Least Square is a method used in Linear Regression to find the best-fit line for a dataset.**

**It does this by calculating the coefficients(Beta) that minimize the sum of squared differences between actual and predicted values.**

```python
# 8:07 AM.


```

```python
df
```

```python
new_X = df[['YearsExperience']]  # x in s 2d array
new_Y = df['Salary'].values  # Ensure y is in 1d array format
```

```python
new_Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(new_X,new_Y,test_size=0.1,random_state=0)
```

```python
x_train_ols  = np.hstack((np.ones((x_train.shape[0],1)),x_train))
x_test_ols  = np.hstack((np.ones((x_test.shape[0],1)),x_test))
x_test_ols         #instead of just [x] the matrix became [1,x]
#  y^ = B0 + B1x1 + B2x2 + .... BnXn
# B0 -> Intercept (bias term)
# b1,b2 -> Slope coeff....
# We add the column of ones so that the OLS matrix formula can estimate the intercept
# (B0) just like any other coefficent. Without it, the regression would have no intercept
# and could only tilt around the origin.
```

```python
x_train_ols
```

$$
{\beta} = (X^T X)^{-1} X^T y
$$

```python
# Compute the OLS coefficents manually
x_transpose = x_train_ols.T
x_transpose
beta = np.linalg.inv(x_transpose @ x_train_ols) @ x_transpose @ y_train
```

```python
print('Intercept:',beta[0])
print('Slope:',beta[1])
```

```python
y_pred_ols = x_test_ols @ beta
```

```python
y_pred_ols
```

```python
r2_score(y_test,y_pred_ols)
```

```python
sns.regplot(x=y_pred_ols,y=y_test)
plt.title('Best Fit Line using OLS')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()
```

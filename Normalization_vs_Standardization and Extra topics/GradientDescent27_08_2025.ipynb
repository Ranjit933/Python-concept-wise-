```python
# Goal of Gradient Descent is to minimize the error.


di = {
    'X' : [-0.712, 0.451, -0.044, 0.753],
    'Y' : [7.217,16.046, -27.837, 144.366]
}
```

```python
import numpy as np
```

```python
import pandas as pd
df = pd.DataFrame(di)
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['X'],df['Y'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[['X']]
y = df[['Y']]
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_  # Value of m
```

```python
reg.intercept_
```

```python
reg.intercept_  #Value of B
#
first_pred = reg.predict(X)
```

```python
plt.scatter(X,y)
plt.plot(X,first_pred,color='red')
```

```python
# Let's apply gradient descent assuming slope is constant m = 78.40
# Let's assume the starting point of intercept b = 0

y_pred = ((78.40 * X) + 0)
```

```python
plt.scatter(X,y)
plt.plot(X,first_pred,color='red',label='OLS')
plt.plot(X,y_pred,color='yellow',label='b=0')
plt.legend()
plt.show()
```

```python
X_col = X.iloc[:,0].values
X_col
```

```python
y_col = y.iloc[:,0].values
y_col
```

```python
m = 78.40
b = 0
loss_slope = -2 * np.sum(y_col - m * X_col - b)
loss_slope
```

```python
lr = 0.1   #Learning rate, which controls how big a step we take while updating the paratemers
step_size = loss_slope * lr
step_size
```

```python
b = b - step_size
b #New Intercept
```

```python
#Let's make new predictions
y_pred1 = ((78.40 * X) + b)
y_pred1
```

```python
plt.scatter(X,y)
plt.plot(X,first_pred,color='red',label='OLS')
plt.plot(X,y_pred,color='yellow',label='b=0')
plt.plot(X,y_pred1, color='green',label='b=20.93')
plt.legend()
plt.show()
```

```python
# 2nd iteration
loss_slope = -2 * np.sum(y_col - m * X_col - b)
loss_slope
```

```python
step_size = loss_slope * lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.40 * X) + b)
y_pred2
```

```python
plt.scatter(X,y)
plt.plot(X,first_pred,color='red',label='OLS')
plt.plot(X,y_pred,color='yellow',label='b=0')
plt.plot(X,y_pred1, color='green',label='b=20.93')
plt.plot(X,y_pred2, color='teal',label='b=25.12')
plt.legend()
plt.show()
```

```python
# 3rd iteration
loss_slope = -2 * np.sum(y_col - m * X_col - b)
loss_slope
```

```python
step_size = loss_slope * lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = ((78.40 * X) + b)
y_pred3
```

```python
plt.scatter(X,y)
plt.plot(X,first_pred,color='red',label='OLS')
plt.plot(X,y_pred,color='yellow',label='b=0')
plt.plot(X,y_pred1, color='green',label='b=20.93')
plt.plot(X,y_pred2, color='teal',label='b=25.12')
plt.plot(X,y_pred3,color='cyan',label='b=25.95')
plt.legend()
plt.show()
```

```python
b = -100
m = 78.40
lr = 0.1
for x in range(10):
    loss_slope = -2 * np.sum(y_col - m * X_col - b)
    b = b - (lr * loss_slope)
    y_pred = m*X_col + b
    plt.plot(X_col,y_pred)

plt.scatter(X,y)
```

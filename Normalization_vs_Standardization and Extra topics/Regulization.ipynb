```python
import pandas as pd
import numpy as np
```

```python
x=pd.DataFrame({
    'Area':[800,900,1000,1100,1200],
    'Bedroom':[1,2,2,3,3],
    'Age':[20,15,10,5,2]
})
```

```python
y=[50,60,65,80,90]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
print('Linear Regression Cooefficients',lr.coef_)
```

# **Ridge Regression:**

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
ridge=Ridge(alpha=1)
ridge.fit(x_train,y_train)
```

```python
print('Ridge Coefficients',ridge.coef_)
```

# **Lasso Regression:**

```python
lasso=Lasso(alpha=1)
lasso.fit(x_train,y_train)
```

```python
print('Lasso Coeficients',lasso.coef_)
```

```python
pred_lr=lr.predict(x_test)
```

```python
ridge_pr=ridge.predict(x_test)
```

```python
lasso_pr=lasso.predict(x_test)
```

```python
from sklearn.metrics import *
```

```python
print('LR MSE:',mean_squared_error(y_test,pred_lr))
```

```python
print('Ridge MSE:',mean_squared_error(y_test,ridge_pr))
```

```python
print('Lasso MSE:',mean_squared_error(y_test,lasso_pr))
```

# **ROC-AUC Curve:**

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(
    n_samples=1000,
    n_features=5,
    n_classes=2,
    random_state=42
)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_prob=lr.predict_proba(x_test)[:,1]
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
fpr,tpr,thresholds=roc_curve(y_test,y_prob)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8))
plt.plot(fpr,tpr)
plt.plot([0,1],[0,1],linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

```python
auc_score=roc_auc_score(y_test,y_prob)
```
